## **1. Tecnologías a utilizar**
---

### **1.1. Selección de tecnologías**
---

La selección tecnológica para este proyecto responde a la necesidad de construir una arquitectura de Big Data genuina, capaz de modelar la complejidad de la rotación de empleados desde varias perspectivas de análisis y almacenamiento. La decisión de integrar diferentes paradigmas de bases de datos y motores de procesamiento surge no solo del interés académico por explorar, sino de la voluntad de simular un entorno profesional realista, donde cada tecnología responde a un problema concreto del dominio de Recursos Humanos.

Las tecnologías seleccionadas son:

- Neon (PostgreSQL en la nube)

- AstraDB (Cassandra en la nube)

- MongoDB Atlas (documental en la nube)

- Neo4j Aura (base de grafos en la nube)

- Apache Spark (Databricks Community Edition) o Dask (Coiled Cloud Free)

La razón de elegir estas tecnologías, por encima de alternativas tradicionales, radica en su especialización y complementariedad. PostgreSQL, en su versión gestionada a través de Neon, es la opción más robusta y estandarizada para almacenar datos estructurados y ejecutar consultas SQL avanzadas, alineándose con la base central de Employee Attrition y facilitando la trazabilidad y consistencia de la información. Sin embargo, una única base relacional resulta insuficiente para cubrir la diversidad de preguntas que plantea la analítica moderna de recursos humanos.

AstraDB, basado en Cassandra, se incorpora por su capacidad probada de gestionar grandes volúmenes de datos y responder con agilidad a consultas de KPIs agregados. En este caso, se utilizará específicamente para mantener actualizadas dos métricas clave: la rotación de empleados por área y la rotación mensual global. Estas vistas agregadas permiten alimentar dashboards de consulta rápida y soportar decisiones ejecutivas en tiempo real, algo para lo que Cassandra ha sido históricamente la opción predilecta en sectores como telecomunicaciones, banca y análisis operacional.

MongoDB Atlas se justifica por su flexibilidad y eficacia para almacenar datos no estructurados. En este proyecto, se convierte en el repositorio natural de los comentarios de retiro, permitiendo analizar, clasificar y eventualmente explorar el sentimiento detrás de las salidas de empleados. Mientras que una base SQL podría forzar la estructura de estos comentarios, MongoDB permite una evolución natural del esquema y se alinea con los procesos reales donde el feedback de empleados es altamente variable y valioso para la gestión de talento.

Neo4j Aura se emplea para modelar y analizar las relaciones organizacionales, especialmente jerarquías y dependencias entre empleados y líderes. El uso de grafos es crucial para detectar patrones de contagio en la rotación, identificar equipos vulnerables y visualizar dinámicas internas que no serían evidentes en un modelo tabular. Esta perspectiva multi-relacional responde a una tendencia creciente en analítica organizacional avanzada, donde la topología de las conexiones importa tanto como los atributos individuales.

Finalmente, Apache Spark (o Dask, según recursos) funge como el motor de procesamiento distribuido y orquestador del flujo de datos. Permite diseñar pipelines ETL robustos, alimentar cada base de datos, generar los KPIs requeridos y, como fase avanzada, entrenar un modelo de machine learning simple para predecir el riesgo de rotación. Esta integración no solo ilustra la capacidad técnica del entorno, sino que responde a la demanda de automatización y análisis predictivo que distingue la ciencia de datos moderna.

Las ventajas de esta arquitectura son múltiples:

- Permite el almacenamiento y análisis especializado para cada tipo de dato (estructurado, agregado, no estructurado y relacional).

- Aprovecha servicios gratuitos y escalables, facilitando el aprendizaje y la replicabilidad.

- Refleja escenarios reales de integración tecnológica en empresas con sistemas heterogéneos.

Entre las principales limitaciones destacan la complejidad en la orquestación de múltiples servicios, las restricciones de los planes gratuitos (capacidad, concurrencia, límites de uso), y la necesidad de generar ciertos datos derivados o simulados (como comentarios de retiro o relaciones jerárquicas) para cubrir todos los casos de uso, dado que el dataset original no incluye todas las dimensiones necesarias.

**Limpieza y preprocesamiento básico**

```python
# Ejemplo: Mapeo de 'Attrition' a valores booleanos
df['Attrition'] = df['Attrition'].map({'Yes': True, 'No': False})

# Validación de nulos
print("Valores nulos por columna:")
print(df.isnull().sum())
```

**Conexión a Postgres (Neon)**

```python
import psycopg2

# Reemplaza con tus datos de acceso a Neon/Postgres
conn = psycopg2.connect("")
cur = conn.cursor()
```

**Creación de tabla**

```python
# Esquema acorde a las columnas del dataset
cur.execute("""
    CREATE TABLE IF NOT EXISTS employee_attrition (
        Employee_ID INTEGER PRIMARY KEY,
        Age INTEGER,
        Gender VARCHAR(10),
        Marital_Status VARCHAR(10),
        Department VARCHAR(50),
        Job_Role VARCHAR(50),
        Job_Level INTEGER,
        Monthly_Income FLOAT,
        Hourly_Rate FLOAT,
        Years_at_Company INTEGER,
        Years_in_Current_Role INTEGER,
        Years_Since_Last_Promotion INTEGER,
        Work_Life_Balance INTEGER,
        Job_Satisfaction INTEGER,
        Performance_Rating INTEGER,
        Training_Hours_Last_Year INTEGER,
        Overtime VARCHAR(5),
        Project_Count INTEGER,
        Average_Hours_Worked_Per_Week FLOAT,
        Absenteeism INTEGER,
        Work_Environment_Satisfaction INTEGER,
        Relationship_with_Manager INTEGER,
        Job_Involvement INTEGER,
        Distance_From_Home FLOAT,
        Number_of_Companies_Worked INTEGER,
        Attrition BOOLEAN
    );
""")
conn.commit()
```

**Inserción de los datos**

```python
# Inserción por lotes. Evita duplicados usando ON CONFLICT.
for row in df.itertuples(index=False):
    cur.execute("""
        INSERT INTO employee_attrition VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON CONFLICT (Employee_ID) DO NOTHING;
    """, row)
conn.commit()
```

**Cassandra – KPIs de rotación por área y por mes**

Razonamiento:
En la práctica, Recursos Humanos necesita monitorear de forma ágil la rotación por áreas y en ventanas temporales (por mes/año). Cassandra, por su arquitectura, está diseñada para responder rápidamente a queries agregadas sobre grandes volúmenes de datos.

Estrategia:

Rotación por área: calcular para cada departamento el porcentaje de empleados que se han retirado (Attrition=True).

Rotación por mes: requiere crear una columna de "fecha de retiro", la cual puede simularse razonablemente (por ejemplo, asignando aleatoriamente dentro del año de salida o usando la fecha de última actualización).

```python
from datetime import datetime, timedelta
import numpy as np
import pandas as pd

# Supongamos que df es el DataFrame principal ya cargado y limpiado

# 1. Rotación por área
kpi_area = df.groupby('Department')['Attrition'].mean().reset_index()
kpi_area.columns = ['Department', 'Attrition_Rate']
print("KPI de rotación por área:\n", kpi_area.head())

# 2. Rotación por mes (simulación de fechas de retiro para quienes se fueron)
# Si el dataset tiene una columna de retiro, úsala; si no, simula:
df['Retirement_Date'] = np.where(df['Attrition'],
    pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 365, df.shape[0]), unit='D'),
    pd.NaT
)

# Filtra el DataFrame para incluir solo las filas donde Retirement_Date no es NaT
df_attrited = df[df['Retirement_Date'].notna()].copy()

df_attrited['Retirement_Date'] = pd.to_datetime(df_attrited['Retirement_Date'])


df_attrited['Year'] = df_attrited['Retirement_Date'].dt.year
df_attrited['Month'] = df_attrited['Retirement_Date'].dt.month

# 1. Rotación por área y año
kpi_area = (df_attrited[df_attrited['Attrition']]
            .groupby(['Department', 'Year'])
            .size()
            .reset_index(name='Attrition_Count'))

# 2. Rotación por mes y año
kpi_mes = (df_attrited[df_attrited['Attrition']]
           .groupby(['Year', 'Month'])
           .size()
           .reset_index(name='Attrition_Count'))

print("KPIs por área/año:\n", kpi_area.head())
print("KPIs por mes/año:\n", kpi_mes.head())
```

Estos KPIs se exportan como tablas y se cargan en Cassandra, optimizados para consultas por partición (Department o Retirement_Month).

**MongoDB Atlas – Comentarios de salida simulados**

Razonamiento:
La realidad en RRHH es que pocas empresas cuentan con feedback estructurado para cada salida. Se generan comentarios simulados basados en reglas fundadas en las variables más influyentes (Job_Satisfaction, Performance_Rating, Attrition, etc.), asegurando que sean coherentes y útiles para análisis de sentimiento o tópico.

Estrategia:

Crear una lista de frases tipo para comentarios de retiro, asignándolas según reglas lógicas.

Ejemplo:

Job_Satisfaction <= 2 → comentario negativo

Job_Satisfaction >= 4 → comentario positivo

Otros casos → comentario neutro

```python
comentarios = {
    'neg': [
        "Me sentía estancado y sin oportunidades de crecer.",
        "La sobrecarga de trabajo y la presión me llevaron a buscar otro empleo.",
        "No me sentía valorado en el equipo."
    ],
    'pos': [
        "Fue una buena experiencia, pero quiero nuevos retos.",
        "Agradezco lo aprendido, pero busco mejores beneficios.",
        "Mi salida fue amistosa y me llevo grandes recuerdos."
    ],
    'neu': [
        "Decisión personal sin mayores motivos.",
        "No tengo comentarios relevantes sobre mi retiro.",
        "Transición normal según mi plan de carrera."
    ]
}

def asignar_comentario(row):
    if not row['Attrition']:
        return None
    elif row['Job_Satisfaction'] <= 2:
        return np.random.choice(comentarios['neg'])
    elif row['Job_Satisfaction'] >= 4:
        return np.random.choice(comentarios['pos'])
    else:
        return np.random.choice(comentarios['neu'])

df['Exit_Comment'] = df.apply(asignar_comentario, axis=1)
print(df[['Employee_ID', 'Job_Satisfaction', 'Exit_Comment']].dropna().head())
```

Solo los empleados con Attrition=True tendrán comentario.

Este campo es fundamental para futuras pruebas de análisis de sentimiento en MongoDB.

**Neo4j Aura — Relaciones jerárquicas**

```python
from neo4j import GraphDatabase

URI = ""
AUTH = ("")

driver = GraphDatabase.driver(URI, auth=AUTH)
driver.verify_connectivity()

def crear_nodos_y_relaciones(tx, empleados, jefes):
    for jefe in jefes:
        tx.run(
            "MERGE (m:Manager {Manager_ID: $Manager_ID, Department: $Department})",
            Manager_ID=jefe['Manager_ID'], Department=jefe['Department'])
    for _, row in empleados.iterrows():
        tx.run(
            """
            MERGE (e:Empleado {Employee_ID: $eid, Job_Role: $role})
            MERGE (m:Manager {Manager_ID: $mid})
            MERGE (e)-[:REPORTS_TO]->(m)
            """,
            eid=int(row['Employee_ID']), role=row['Job_Role'], mid=row['Manager_ID'])

with driver.session() as session:
    session.write_transaction(crear_nodos_y_relaciones, empleados, jefes)
driver.close()
```

## Paso 1: Ingesta y limpieza de datos desde Postgres

**Descripción:**  
Leemos la tabla `employee_attrition` desde el servicio Neon/Postgres, la reparticionamos para paralelismo y aplicamos operaciones de limpieza básica:

- Eliminación de duplicados por `Employee_ID`.  
- Imputación de valores nulos en columnas numéricas (`Monthly_Income`, `Hourly_Rate`, `Absenteeism`).  
- Casting de la columna `Attrition` a booleano.  
- Flag `data_quality_issue` para detectar filas con ingresos o años negativos.

---

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import json

# 1) Crear sesión de Spark con conector Postgres y Cassandra
# 1) Construye la sesión Spark incluyendo el paquete del conector

spark = (SparkSession.builder
    .appName("Attrition_KPIs_Spark_Cassandra")
    .config("spark.jars.packages",
                 ",".join([
                   "org.postgresql:postgresql:42.2.18",
                 ]))
    .getOrCreate())

# 2) Leer desde Neon/Postgres vía JDBC
PG_URL = ""
PG_PROPS = {"user": "", "password": "", "sslmode": "require", "driver": "org.postgresql.Driver"}

master_df = (spark.read
             .jdbc(url=PG_URL,
                   table="employee_attrition",
                   properties=PG_PROPS)
             .repartition(8))  # particionamos para paralelismo

print("Registros leídos:", master_df.count())
master_df.printSchema()
```

```python
# 3) Drop duplicados por Employee_ID
df_nodup = master_df.dropDuplicates(["Employee_ID"])

# 4) Imputación de nulos y flags de calidad
df_clean = (df_nodup
    # a) Rellenar nulos numéricos con 0
    .fillna({"Monthly_Income": 0, "Hourly_Rate": 0, "Absenteeism": 0})
    # b) Casting de booleanos
    .withColumn("Attrition", F.col("Attrition").cast("boolean"))
    # d) Flag de problemas de datos
    .withColumn("data_quality_issue",
                (F.col("Monthly_Income") < 0) |
                (F.col("Years_at_Company") < 0) |
                F.col("Attrition").isNull()
               )
)

print("Duplicados removidos:", master_df.count() - df_nodup.count())
print("Registros con issues:", df_clean.filter("data_quality_issue").count())
df_clean.show(5, truncate=False)
```

## Paso 2: Cálculo de KPIs y full‑upsert en Cassandra

**Descripción:**  
A partir del DataFrame limpio (`df_clean`):

1. Simulamos `Retirement_Date` para quienes tengan `Attrition = true`.  
2. Extraemos `Year` y `Month` de esa fecha.  
3. Agrupamos para obtener:  
   - **`kpi_area`**: conteo de retiros por `Department` y `Year`.  
   - **`kpi_mes`**: conteo de retiros por `Year` y `Month`.  
4. Conectamos a AstraDB/Cassandra, vaciamos (`TRUNCATE`) las tablas de KPI y subimos el batch completo.  
5. Cada `INSERT` actúa como **upsert**, garantizando idempotencia.

---


```python
# --- 1. Convierte tu Spark DataFrame limpio a Pandas (si aplica) ---
# Si ya trabajas en Pandas, ignora esta línea:
df = df_clean.toPandas()

import numpy as np
import pandas as pd

# 1. Rotación por área
kpi_area = df.groupby('department')['Attrition'].mean().reset_index()
kpi_area.columns = ['Department', 'Attrition_Rate']
print("KPI de rotación por área:\n", kpi_area.head())

# 2. Rotación por mes (simulación de fechas de retiro para quienes se fueron)
# Si el dataset tiene una columna de retiro, úsala; si no, simula:
df['Retirement_Date'] = np.where(df['Attrition'],
    pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 365, df.shape[0]), unit='D'),
    pd.NaT
)

# Filtra el DataFrame para incluir solo las filas donde Retirement_Date no es NaT
df_attrited = df[df['Retirement_Date'].notna()].copy()

df_attrited['Retirement_Date'] = pd.to_datetime(df_attrited['Retirement_Date'])


df_attrited['Year'] = df_attrited['Retirement_Date'].dt.year
df_attrited['Month'] = df_attrited['Retirement_Date'].dt.month

# 1. Rotación por área y año
kpi_area = (df_attrited[df_attrited['Attrition']]
            .groupby(['department', 'Year'])
            .size()
            .reset_index(name='Attrition_Count'))

# 2. Rotación por mes y año
kpi_mes = (df_attrited[df_attrited['Attrition']]
           .groupby(['Year', 'Month'])
           .size()
           .reset_index(name='Attrition_Count'))

print("KPIs por área/año:\n", kpi_area.head())
print("KPIs por mes/año:\n", kpi_mes.head())
```

```python
# 1) Conexión a AstraDB (Cassandra)
from cassandra.cluster import Cluster
from cassandra.auth import PlainTextAuthProvider
import json

cloud_config = {"secure_connect_bundle": "bundle.zip"}
with open('token.json') as f:
    token = json.load(f)

# Use the correct protocol version
auth_provider = PlainTextAuthProvider(token["clientId"], token["secret"])
cluster       = Cluster(cloud=cloud_config, auth_provider=auth_provider, protocol_version=4)
session       = cluster.connect()

# 2) (Opcional) Full‐refresh: vacía las tablas de KPI
session.execute("TRUNCATE test.kpi_rotacion_area;")
session.execute("TRUNCATE test.kpi_rotacion_mes;")

# 3) Inserción/upsert de KPIs por área/año
# Filter out rows with NaN values in 'Year' or 'Attrition_Count'
kpi_area_clean = kpi_area.dropna(subset=['Year', 'Attrition_Count'])
for _, row in kpi_area_clean.iterrows():
    session.execute("""
        INSERT INTO test.kpi_rotacion_area (Department, Year, Attrition_Count)
        VALUES (%s, %s, %s)
    """, (row['department'], int(row['Year']), int(row['Attrition_Count'])))

# 4) Inserción/upsert de KPIs por mes/año
# Filter out rows with NaN values in 'Year', 'Month', or 'Attrition_Count'
kpi_mes_clean = kpi_mes.dropna(subset=['Year', 'Month', 'Attrition_Count'])
for _, row in kpi_mes_clean.iterrows():
    session.execute("""
        INSERT INTO test.kpi_rotacion_mes (Year, Month, Attrition_Count)
        VALUES (%s, %s, %s)
    """, (int(row['Year']), int(row['Month']), int(row['Attrition_Count'])))


print("✔ KPIs cargados en Cassandra (full‐upsert).")
```

## Paso 3: Simulación de comentarios y full‑refresh en MongoDB

**Descripción:**  
Generamos de forma determinista un comentario de salida para cada empleado con `Attrition = true`, basándonos en su `Employee_ID` y `Job_Satisfaction`.  
Luego:

## Paso 4: Reconstrucción del grafo organizacional en Neo4j Aura

**Descripción:**  
Creamos nodos y relaciones para modelar la estructura jerárquica:

1. Generamos una lista de jefes (`Manager`) única por `Department`.  
2. Asignamos a cada empleado un `Manager_ID` basado en su departamento.  
3. Conectamos a Neo4j Aura y ejecutamos un **full-refresh** del grafo:  
   - `MATCH (n) DETACH DELETE n`  
   - `MERGE` de nodos `Manager` y `Empleado`.  
   - `MERGE` de relaciones `[:REPORTS_TO]`.  
4. El uso de `MERGE` asegura idempotencia ante reintentos.

---

```python
from neo4j import GraphDatabase

# 1) Prepara listas de jefes y empleados
departments = df['department'].unique()
jefes = [{'Manager_ID': f"{dept}_mgr", 'Department': dept} for dept in departments]

# Agrega columna Manager_ID al DataFrame
df['Manager_ID'] = df['department'].apply(lambda x: f"{x}_mgr")
empleados = df[['employee_id', 'job_role', 'Manager_ID']].to_dict('records')

# 2) Conecta a Neo4j Aura
URI  = ""
AUTH = ("")
driver = GraphDatabase.driver(URI, auth=AUTH)
driver.verify_connectivity()

# 3) Función para full-refresh del grafo
def rebuild_graph(tx, jefes, empleados):
    # a) Borra todo el grafo actual
    tx.run("MATCH (n) DETACH DELETE n")
    # b) Crea nodos Manager
    for jefe in jefes:
        tx.run(
            "MERGE (m:Manager {Manager_ID: $Manager_ID, Department: $Department})",
            **jefe
        )
    # c) Crea nodos Empleado y relaciones REPORTS_TO
    for emp in empleados:
        tx.run(
            """
            MERGE (e:Empleado {Employee_ID: $employee_id})
            SET e.Job_Role = $job_role
            WITH e
            MATCH (m:Manager {Manager_ID: $Manager_ID})
            MERGE (e)-[:REPORTS_TO]->(m)
            """,
            **emp
        )

# 4) Ejecuta el rebuild
with driver.session() as session:
    session.write_transaction(rebuild_graph, jefes, empleados)

driver.close()
print("✅ Grafo Neo4j reconstruido (full-refresh).")
```

## Paso 5: Entrenamiento y evaluación ML con Spark MLlib

**Descripción:**  
Usamos el mismo DataFrame limpio de Postgres para:

1. Indexar la variable categórica `Overtime` y la etiqueta `Attrition`.  
2. Armar un vector de features con `VectorAssembler` (edad, ingresos, antigüedad, Work_Life_Balance, Job_Satisfaction, Overtime_Idx).  
3. Entrenar un `DecisionTreeClassifier` (maxDepth=4) con un split 70/30.  
4. Evaluar con **Accuracy** y **F1 score**.  
5. Guardar las predicciones en la tabla `attrition_scores` de Postgres para ser consumidas por dashboards o alertas.

---

```python
from pyspark.sql.functions import col, udf
from pyspark.sql.types import DoubleType
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.ml.linalg import Vector

# Cast Attrition to integer before using in the pipeline
df_ml = df_clean.withColumn("Attrition_int", F.col("Attrition").cast("integer"))

# 3) Indexa la variable Overtime (“Yes”/“No”) y la etiqueta Attrition
si_overtime = StringIndexer(inputCol="overtime", outputCol="overtime_Idx", handleInvalid="skip")
si_label    = StringIndexer(inputCol="Attrition_int",  outputCol="label",       handleInvalid="skip")

# 4) Prepara el vector de características
feature_cols = [
    "age", "monthly_income", "years_at_company",
    "work_life_balance", "job_satisfaction", "overtime_Idx"
]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features", handleInvalid="keep")

# 5) Clasificador: Árbol de decisión
dt = DecisionTreeClassifier(labelCol="label", featuresCol="features", maxDepth=4)

# 6) Construye el pipeline
pipeline = Pipeline(stages=[si_overtime, si_label, assembler, dt])

# 7) Divide en train/test
train_df, test_df = df_ml.randomSplit([0.7, 0.3], seed=42)

# 8) Entrena el modelo
model = pipeline.fit(train_df)

# 9) Predice sobre test
pred = model.transform(test_df)

# Define a UDF to extract the probability of attrition (class 1)
get_attrition_probability = udf(lambda v: float(v[1]) if v is not None and len(v) > 1 else None, DoubleType())

# 10) Evaluate performance
evaluator_multi = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

acc  = evaluator_multi.setMetricName("accuracy").evaluate(pred)
f1   = evaluator_multi.setMetricName("f1").evaluate(pred)

print(f"▶️ Accuracy : {acc:.3f}")
print(f"▶️ F1 score : {f1:.3f}")

# 11) Guarda los scores en Postgres para consultas futuras
pred.select("employee_id", get_attrition_probability("probability").alias("attrition_probability"), "prediction") \
    .withColumnRenamed("prediction", "predicted_attrition") \
    .write \
    .mode("overwrite") \
    .jdbc(url=PG_URL, table="attrition_scores", properties=PG_PROPS)

print("✅ Predicciones guardadas en Postgres (tabla attrition_scores).")
```
